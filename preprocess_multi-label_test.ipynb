```python
unique_labels = []
for idx in range(len(raw_df_test)):
    row = raw_df_test.iloc[idx]
    unique_labels.extend(row["labels"].split())

unique_labels = list(set(unique_labels))
```

```python
raw_df_test
```

```python
raw_df_test["multi_one_hot_label"] = ""
```

```python
pseudo_collumn = []
for idx in range(len(raw_df_test)):
    label = list(raw_df_test.iloc[idx]["labels"].split())
    multi_one_hot_label = [0] * len(unique_labels)

    for x in label:
        multi_one_hot_label[unique_labels.index(x)] = 1
    
    assert len(label) == sum(multi_one_hot_label)

    raw_df_test.iloc[idx]["multi_one_hot_label"] = multi_one_hot_label

    print(multi_one_hot_label, "    ==  ", raw_df_test.iloc[idx]["multi_one_hot_label"])
    # if idx > 4:
    #     break

    pseudo_collumn.append(multi_one_hot_label)
```

```python
raw_df_test["multi_one_hot_label"] = pseudo_collumn
```
